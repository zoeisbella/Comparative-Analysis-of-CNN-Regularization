```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np

# 检查 TensorFlow 版本
print("TensorFlow 版本:", tf.__version__)

# 加载和准备 CIFAR-10 数据集
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images = train_images / 255.0
test_images = test_images / 255.0

# 定义图像增强生成器
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# 基础模型函数
def create_model(use_normalization=False, use_dropout=False):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3), padding='same'))
    if use_normalization:
        model.add(tf.keras.layers.GroupNormalization(groups=32, axis=3))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(8, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((4, 4)))
    if use_normalization:
        model.add(tf.keras.layers.GroupNormalization(groups=2, axis=3))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    if use_dropout:
        model.add(layers.Dropout(0.5))
    model.add(layers.Dense(10, activation='softmax'))
    return model

# 训练函数
def train_model(model, use_augmentation=False, epochs=5):
    model.compile(optimizer='adam',
                 loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 metrics=['accuracy'])
    if use_augmentation:
        history = model.fit(datagen.flow(train_images, train_labels, batch_size=32),
                          epochs=epochs,
                          validation_data=(test_images, test_labels))
    else:
        history = model.fit(train_images, train_labels,
                          batch_size=32,
                          epochs=epochs,
                          validation_data=(test_images, test_labels))
    return history

# 定义实验配置（8种组合）
experiments = {
    'Baseline': {'augmentation': False, 'normalization': False, 'dropout': False},
    'Augmentation': {'augmentation': True, 'normalization': False, 'dropout': False},
    'Dropout': {'augmentation': False, 'normalization': False, 'dropout': True},
    'Normalization': {'augmentation': False, 'normalization': True, 'dropout': False},
    'Aug + Dropout': {'augmentation': True, 'normalization': False, 'dropout': True},
    'Aug + Norm': {'augmentation': True, 'normalization': True, 'dropout': False},
    'Dropout + Norm': {'augmentation': False, 'normalization': True, 'dropout': True},
    'Aug + Dropout + Norm': {'augmentation': True, 'normalization': True, 'dropout': True}
}

# 运行实验并存储结果
results = {}
for exp_name, config in experiments.items():
    print(f"\n运行实验: {exp_name}")
    try:
        model = create_model(
            use_normalization=config['normalization'],
            use_dropout=config['dropout']
        )
        history = train_model(
            model,
            use_augmentation=config['augmentation']
        )
        results[exp_name] = {
            'history': history.history,
            'test_acc': history.history['val_accuracy'][-1]
        }
    except Exception as e:
        print(f"实验 {exp_name} 失败: {str(e)}")

# 绘制折线图
if results:
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    for exp_name, result in results.items():
        plt.plot(result['history']['accuracy'], label=exp_name)
    plt.title('训练准确率对比')
    plt.xlabel('轮次')
    plt.ylabel('准确率')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    for exp_name, result in results.items():
        plt.plot(result['history']['val_accuracy'], label=exp_name)
    plt.title('验证准确率对比')
    plt.xlabel('轮次')
    plt.ylabel('准确率')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # 打印最终测试准确率
    print("\n最终测试准确率结果:")
    for exp_name, result in results.items():
        print(f"{exp_name}: {result['test_acc']:.4f}")
else:
  print("没有成功完成任何实验，无法生成图表")
```
